```python
# Dependencies
import requests
import json
import pandas as pd
import numpy as np
```

```python
# URL for GET requests to retrieve data for Covid 19 cases from centre for diesase control and prevention
url = "https://data.cdc.gov/resource/n8mc-b4w4.json?res_state=NJ&$limit=20000"
```

```python
# Print the response object to the console
response = requests.get(url)
print( response.url)
```

```python
# New Jersey Data 

data = response.json()

data_df = pd.DataFrame(data)

data_df
```

```python
# New Jersey data for hospitalisation 
df_hosp = data_df['hosp_yn'].value_counts()
df_hosp
```

```python
# No of cases 
no_cases = data_df["res_state"].count()
df_cases = pd.DataFrame({"Total no. of cases in sample":[no_cases]})
df_cases
```
